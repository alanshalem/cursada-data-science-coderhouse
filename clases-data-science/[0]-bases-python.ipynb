```python
# Operadores de comparacion (Dan resultados logicos, los booleanos)

x = 10
y = 20

# Igualdad
print(x == y)

# Desigualdad 
print(x != y)

# Mayor que 
print(x > y)

# Menor que
print(x < y)

# Mayor o igual que
print(x >= y)

```

```python
# Operadores de identidad o pertenecia

x = 10
y = 20

# X es lo mismo que Y?
print(x is y)

# X no es lo mismo que Y?
print(x is not y)


y = [2, 7, 10]


# Esta X contenido en Y?
print(x in y)

# No esta contenido X en Y?
print(x not in y)
```
