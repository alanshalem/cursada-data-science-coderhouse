# Condicionales if
Nos sirven para preguntar si se cumple una condicion y en relacion a eso hacer una de diferentes opciones.
    

```python
edadIngresada = int(input('Ingrese su edad: '))

if edadIngresada >= 18:
    print('Usted es mayor de edad')
else:
    print('Usted es menor de edad')
```

# Bucles 'while'
Sirven para repetir instrucciones cuando no sabemos la cantidad de veces a hacerlo

```python
#Casi siempre necesitaremos que la condicion cambie dentro del bucle (el 90% de las veces)

i = 0
while i < 5:
    i += 1
    print(i)
```

```python
alumnos = 5
listado=[]

while alumnos != 0:
    alumno = input('Ingrese el nombre del alumno: ')
    listado.append(alumno)
    alumnos -= 1
    
print(listado)
```

```python
gente = 'y'
while gente == 'y':
    edadIngresada = int(input('Ingrese su edad: '))
    if edadIngresada >= 18:
        print('Usted es mayor de edad')
        lugar = input('Para donde queres ir? Al baño, a la pista o a la barra?:')
        if lugar == 'baño':
            print('Vamos para el baño')
        elif lugar == 'pista':
            print('Vamos para la pista')
        elif lugar == 'barra':
            print('Vamos para la barra')
        else:
            print('Pusiste cualquier cosa')
    else:
        print('Usted es menor de edad, no puede pasar')
    gente = input('Hay mas gente? (y/n) :')

print('\nNo hay mas gente, noche terminada.')
```

# Elementos extras

Tenemos varios elementos que podemos usar en los bucles: el pass, continue y el break

El pass puede utilizar tambien en el if


```python
numero = 2

if numero == 2:
    pass

while numero == 2:
    pass

for i in range(0, numero):
    pass
```

```python
for i in range(0, 10):
    nombre = input('Ingrese su nombre: ')
    if nombre == 'luca':
        print(f'hola profe {nombre}')
        break
    else:
        print(f'Hola alumno {nombre}')
```

```python
contador = 0        
listado = ['francisco','luca', 'juan', 'carlos', 'felipe', 'matias', 'juanita', 'maria', 'jorge', 'pepe', 'octavio', 'victoria']
for nombre in listado:
    if nombre == 'luca' or nombre == 'juan' or nombre == 'jorge' or nombre == 'octavio':
        continue
    contador += 1
        
print(f'En el curso hay {contador} alumnos ')
```

```python
listado = []
gente = 'y'
while gente == 'y':
    edadIngresada = int(input('Ingrese su edad: '))
    if edadIngresada >= 18:
        nombre = input('ingrese su nombre: ')
        listado.append(nombre)
        print('Usted es mayor de edad')
        lugar = input(F'A donde vas {nombre}? (barra, baño o pista): ')
        if lugar == 'baño':
            print('Vamos para el baño')
        elif lugar == 'pista':
            print('Vamos para la pista')
        elif lugar == 'barra':
            print('Vamos para la barra')
    else:
        print('Usted es menor de edad, no puede pasar')
    gente = input('Hay mas gente? (y/n) :')
    
print('\nNo hay mas gente, noche terminada.')

for persona in listado:
    print(f'Hoy vino {persona}')
    
```

# Funciones
El codigo se suele divir en funciones, para que sea mas reutilizable y facil de leer.
    
    

```python
def suma(a, b):
    c = a + b
    return c

resultado = suma(10, 8)
print(resultado)
```

```python
def multiplicacion(x, y = 2): #Los que tengan valores predeterminados deben ir a la derecha
    z = x * y
    return z

resultado = multiplicacion(5, 3)
print(resultado)

resultado = multiplicacion(10)
print(resultado)
```

```python
def resta(x, y):
    z = x - y
    return(z)

resultado = resta(10, 5)
print(resultado)
```

```python
def lugar():
    lugar = input(F'A donde vas {nombre}? (barra, baño o pista): ')
    if lugar == 'baño':
        print('Vamos para el baño')
    elif lugar == 'pista':
        print('Vamos para la pista')
    elif lugar == 'barra':
        print('Vamos para la barra')


listado = []
gente = 'y'
while gente == 'y':
    edadIngresada = int(input('Ingrese su edad: '))
    if edadIngresada >= 18:
        nombre = input('ingrese su nombre: ')
        listado.append(nombre)
        print('Usted es mayor de edad')
        lugar()
    else:
        print('Usted es menor de edad, no puede pasar')
    gente = input('Hay mas gente? (y/n) :')
    
print('\nNo hay mas gente, noche terminada.\n')

for persona in listado:
    print(f'Hoy vino {persona}')
```

# Tipos de datos

Define que operaciones podemos hacer.

Se dividen en simples y estructurados

Datos simples: La mayoria los explicamos, los vemos desde la diapositiva

Datos estructurados: Tienen mas de un objeto en su interior. Ej:

```python
listas = [1, 2, 3]   # Son de tipo list, estos los vamos a usar un monton, ya que son ordenados

tuplas = (1, 2, 3)   # Son de tipo tuple, Son listas ordenadas inmutbles, no deben ni pueden cambiar a lo largo del programa
```

```python
conjuntos
# Son de tipo set, son listas no ordenas que no pueden contener un elemento mas de una vez, son como conjuntos matematicos.
# los, se les pueden aplicar funciones especificas.

a = {1, 2, 3, 4}
b = {2, 4, 6, 8}
a | b
```

```python
a = {1, 2, 3, 4}
b = {2, 4, 6, 8}
a & b
```

```python
a = {1, 2, 3, 4}
b = {2, 4, 6, 8}
a - b
```

```python
diccionario = {
    'rojo' : 'red',
    'verde' : 'green',
    'azul' : 'blue'
}
for x in diccionario:
    print(f'El color {x} en ingles, se dice {diccionario[x]}')
```
