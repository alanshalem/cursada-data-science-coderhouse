# NumPy
Nos va a dar una estructura de datos muy util y eficiente, los arrays.
Estos pueden contener solo un tipo de dato, lo cual los vuelve muy eficientes (cuidado con enteros y floats)


```python
import numpy as np
```

Hay distintas formas de crearlos

```python
array1 = np.array(range(10))
print(array1)
```

```python
Np_cero = np.zeros(10)
print(Np_cero)
```

```python
Np_cero_int = np.zeros(10, dtype=int)
print(Np_cero_int)
```

```python
Np_uno = np.ones(10)
print(Np_uno)
```

```python
Np_relleno = np.full(10,256)
print(Np_relleno)
```

Podemos armar arrays de 2 dimensiones.

Tambien podemos armarlos con numeros "aleatorios"

```python
arrayBidimensional = np.random.randint(10, size=(3,4))
print(arrayBidimensional)

arrayBidimensional2 = np.random.randint(10, size=(4,6))
print(arrayBidimensional2)
```

```python
print(arrayBidimensional[2,3])
```

# Propiedades de los arrays



```python
# Dimensiones: con .ndim nos devolvera la cantidad de dimensiones.

print(Np_cero.ndim)

print(arrayBidimensional.ndim)
```

```python
# Forma: Podemos ver de que manera se ordenan los datos.

print(arrayBidimensional.shape)

print(arrayBidimensional2.shape)
```

```python
# Tamaño: Podemos la cantidad de elementos de un array, es como el .len de las listas

print(Np_cero_int.size)

print(arrayBidimensional.size)
```

```python
# Tipo de datos: Podemos ver el tipo de dato que tienen nuestros array (recordemos, solo pueden contener elementos de un tipo)

print(Np_cero.dtype)

print(arrayBidimensional.dtype)
```

```python
# Tamaño de elementos: Devuelve el tamaño en bytes de los elementos

print(Np_cero.itemsize)    #En este caso son floats

print(arrayBidimensional.itemsize)     #En este caso son enteros
```

```python
# Tamaño total: Bytes totales consumidos en el array

print(arrayBidimensional2.nbytes)

print(Np_cero_int.nbytes)
```

# Indexado y acceso

Podemos ingresar a los elementos mediante un indice, el cual comienza en 0 y va hasta el ultimo elemento del array.

Podemos acceder a los elementos desde el final con numeros negativos Ej: -1 para el ultimo elemento del array y -2 el penultimo

Para acceder a un elememnto de una matriz debemos indicar fila y columna.


Podemos seleccionar partes del array y separlo del grande. (Debemos marcar desde donde y hasta donde recortar *

    Importante, el 'desde' incluye el valor y el 'hasta' no. (es como si 'desde' fuera <= y 'hasta' > )

```python
arrayPruebas = np.arange(2,20)
arrayPruebas2D = np.random.randint(10, size=(6,8))
print(arrayPruebas)
print(arrayPruebas2D)
```

```python
print(arrayPruebas[9])
print(arrayPruebas[15])
```

```python
print(arrayPruebas[-1])
print(arrayPruebas[-3])
```

```python
print(arrayPruebas2D[2,0])
print(arrayPruebas2D[1,1])
```

```python
print(arrayPruebas2D[1,-2])
print(arrayPruebas2D[1,-1])
```

```python
print(arrayPruebas)
print(arrayPruebas2D)
```

```python
print(arrayPruebas[2:4])
print(arrayPruebas[4:15])
print(arrayPruebas[4:15:2])
```

```python
print(arrayPruebas[:4])
print(arrayPruebas[4:])
print(arrayPruebas[10::2])
```

```python
print(arrayPruebas2D[0:4,1:3])
```

```python
print(arrayPruebas2D[0::2, :3])
```

Reshape

    Como podriamos armar un tablero de ajedrez ?

```python
ajedrez = np.arange(1,65).reshape(8,8)
print(ajedrez)
```

Concatenacion: Nos sirve para "pegar" dos arrays, uno al lado del otro.

    Podemos hacerlo de 3 maneras, con .concatenate, con .vstack y con .hstack

```python
# .concatenate: concatena arrays de una dimension uno a continuacion del otro
Array_1 = np.random.randint(10, size=5)
Array_2 = np.random.randint(10, size=5)
Arrays_concatenados = np.concatenate([Array_1, Array_2])
print(Array_1)
print(Array_2)
print(Arrays_concatenados)
```

```python
# .vstack: concatena arrays de manera vertical

print(ajedrez)

ajedrezX2Vertical=np.vstack([ajedrez, ajedrez])
print(ajedrezX2Vertical)
```

```python
# .hstack: concatena arrays de manera vertical

print(ajedrez)

ajedrezX2Horizontal=np.hstack([ajedrez, ajedrez])
print(ajedrezX2Horizontal)
```

Splitting

    Consiste en dividir los arreglos en partes mas pequeñas

```python
print(Arrays_concatenados)
arrayDividido = np.split(Arrays_concatenados, [3])
print(arrayDividido)
```

```python
print(Arrays_concatenados)
arrayDividido = np.split(Arrays_concatenados, [2,5])
print(arrayDividido)
```

```python
arrayAlmedio = np.split(Arrays_concatenados, 2)
print(arrayAlmedio)
```

```python
primerMitad, segundaMitad = arrayAlmedio
print(primerMitad)
print(segundaMitad)
```

```python
# .vsplit: corta arrays de manera vertical

print(ajedrezX2Vertical)
```

```python
ajedrez1 , ajedrez2=np.vsplit(ajedrezX2Vertical,[8])
print(ajedrez1)
print(ajedrez2)
```

```python
# .hsplit: corta arrays de manera vertical

print(ajedrezX2Horizontal)
```

```python
ajedrez1 , ajedrez2=np.hsplit(ajedrezX2Horizontal,[8])
print(ajedrez1)
print(ajedrez2)
```
