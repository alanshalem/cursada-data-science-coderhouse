# Pandas

Estructuras principales

    Objetos Series: Se construyen a partir de listas o numpy arrays, son muy parecidos, solo que tienen indices propios.

```python
import numpy as np
import pandas as pd
```

```python
numeros = range(0, 5)
numeros_series = pd.Series(numeros)
print(numeros_series)
```

```python
indices = ['primero', 'segundo', 'tercero', 'cuarto', 'quinto']
numeros_series = pd.Series(numeros, index = indices)
print(numeros_series)
```

```python
print(numeros_series.index)
print(numeros_series.values)
```

```python
print(numeros_series['primero'])
print(numeros_series[0])
```

```python
numeros_series['primero'] = 5
print(numeros_series)
```

        Objetos Data Frames: Se los puede pensar como una tabla, algo parecido al los arrays bidimensionales, pero mejores.
        
        Como toda tabla, consta de filas y columnas.
        Cada fila tiene un elemento índice asociado.


```python
modelos = ['A4 3.0 Quattro 4dr manual',
           'A4 3.0 Quattro 4dr auto',
           'A6 3.0 4dr',
           'A6 3.0 Quattro 4dr',
           'A4 3.0 convertible 2dr']
peso = [3583, 3627, 3561, 3880, 3814]
precios = ['$33,430', '$34,480', '$36,640', '$39,640', '$42,490']

autosPeso = pd.Series(peso, index = modelos)
autosPrecios = pd.Series(precios, index = modelos)

print(autosPeso)
print(autosPrecios)
```
