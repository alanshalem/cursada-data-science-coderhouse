# Graficos de barra

Se usan para comparar y poner en perspectiva distintas variables categoricas.

El eje x representa las categorias y la altura de la barra la cantidad de elementos o el valor del mismo.

Se grafican con ax.bar() y recibe minimo, 2 parametros

        Etiquetas para el eje X
        
        La altura de la barra de cada etiqueta

```python
precipitacionesAcumuladas = df_lluvias.sum()
precipitacionesAcumuladas
```

```python
fig, ax = plt.subplots(figsize = (8,4))
ax.bar(df_lluvias.columns, precipitacionesAcumuladas)

ax.set_title('Precipitaciones acumuladas desde 1965 - 2002 segun el mes')
ax.set_ylabel('Precipitacion total (mm)')
ax.set_xlabel('Mes')
```

# Historigrama

La altura de cada barra representa la proporcion o cantidad de los distintos valores de una variable numerica

Requiere clasificar a los datos en intervalos de clase

Se construyen con ax.hist y recibe como parametro:

        Arreglo de valores
        
        Bins, osea, la cantidad de intervalos a construir
        
El grfico de barras: se usa para graficar el numero de veces que se repite algn valor y es para datos no agrupados y estan separadas una barra de otra y lo importante es identificar la barra mas alta.

Histograma: Es para datos agrupados, todas las barras estn juntas, sin dejar espacios de separacin,y se usan para saber el comportamiento de los datos utilizando todas las barras.

```python
df_lluvias.values.flatten()
```

```python
fig, ax = plt.subplots(figsize = (8,4))
ax.hist(df_lluvias.values.flatten(), bins = 10)

ax.set_title = ('Histograma de precipitaciones')
ax.set_xlabel('Intervalos de precipitaciones (mm.)')
ax.set_ylabel('Frecuencia absoluta')
```

# Multiples elementos

Podemos resaltar distintos elementos, para ayudarnos a visualizar mejor los datos.

No debemos pasarnos con la cantidad e informacion o el grafico no sera intuitivo y legible.

```python
fig, ax = plt.subplots(figsize=(12, 3))  
ax.plot(df_lluvias.index, df_lluvias['Jan'], label='Precipitaciones de enero')
ax.plot(df_lluvias.index, df_lluvias['Feb'], label='Precipitaciones de febrero', color='C1')

maximo_enero = df_lluvias['Jan'].max()
maximo_febrero =  df_lluvias['Feb'].max()

promedio_enero = df_lluvias['Jan'].mean()
promedio_febrero = df_lluvias['Feb'].mean()

ax.axhline(maximo_enero, color='red', linestyle='--', alpha=0.5, linewidth=3, label='Mxima de enero')
ax.axhline(maximo_febrero, color='red', linestyle=':', alpha=0.5, linewidth=3, label='Mxima de febrero') #modificar linestyle

ax.axhline(promedio_enero, color='black', linestyle='--', alpha=0.5, linewidth=3, label='Mxima de enero')
ax.axhline(promedio_febrero, color='black', linestyle=':', alpha=0.5, linewidth=3, label='Mxima de febrero')

ax.axvline(1970, color = 'green', linestyle = 'solid', alpha=0.5, linewidth=3, label='Periodo de sequia enero')
ax.axvline(1980, color = 'green', linestyle = 'solid', alpha=0.5, linewidth=3)


ax.set_xlabel('Ao')  
ax.set_ylabel('Precipitacin (mm.)') 
ax.set_title('Precipitaciones de enero y febrero') 	
ax.set_xlim(df_lluvias.index[0], df_lluvias.index[-1])
ax.legend() 

fig.savefig("comparativaEneFeb.pdf")
```
