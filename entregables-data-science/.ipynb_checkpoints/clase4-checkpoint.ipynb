```python
import numpy as np

array_random = np.random.randint(100, size=100)
mediana = np.median(array_random)

print(array_random)
print(mediana)
```

```python
numero_a_calcular=5

def factorial(x: int) -> int:
    if x<0:
        return "Factorial no esta definido para valores negativos"
    else:
      factorial=np.math.factorial(x)
      return factorial

respuesta = factorial(numero_a_calcular)
print(respuesta)
```

```python
numero_inicial_a_calcular = 9
numero_final_a_calcular = 15

def sumar_serie(x: int, y: int):
    if x>y:
        return "No se puede calcular"
    else:
        array = np.arange(x, y+1)
        return np.sum(array)


respuesta = sumar_serie(numero_inicial_a_calcular, numero_final_a_calcular)
print(type(respuesta))
```
