```python
# Escribir una función para calcular el factorial de un número cualquiera.
# SOLUCION CON RECURSIVIDAD
numero_a_calcular = 5


def factorial(numero):  # Aquí definimos una funcion "factorial"
    if numero == 0:
        return 1
    else:
        # Calculamos el factorial del numero mediante recursividad
        return numero*factorial(numero-1)


resultado = factorial(numero_a_calcular)
print(resultado)
```

```python
# Escribir una función para calcular el factorial de un número cualquiera.
# SOLUCION SIN RECURSIVIDAD
numero_a_calcular = 6


def factorial(numero):
    contador = 0
    fact = 1
    if numero == 0:
        return fact
    else:
        for contador in range(1, (numero+1)):
            fact = contador * fact
    return fact


resultado = factorial(numero_a_calcular)
print(resultado)
```

```python
# Escribir una función para calcular la suma de una serie comenzando por un número cualquiera y terminando en otro número que debe ser mayor al primero.

numero_inicial_a_calcular = 9
numero_final_a_calcular = 15


def sumar_serie(numero_inicial, numero_final):
    suma_total = 0
    if(numero_final < numero_inicial):
        return "El numero final es menor al numero inicial"
    else:
        for i in range(numero_inicial, numero_final+1):
            suma_total += i
        return suma_total


resultado = sumar_serie(numero_inicial_a_calcular, numero_final_a_calcular)
print(resultado)
```
